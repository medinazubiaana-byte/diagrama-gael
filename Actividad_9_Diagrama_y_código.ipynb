#### En una tienda en linea, si el valor de la compra es mayor a 800 pesos el envio es grati, pero es menor se le añadira 150 pesos por el coste del envio

```python
# En una tienda en linea, si el valor de la compra es mayor a 800 pesos el envio es gratis, pero si es menor se le añadira 150 pesos por el coste del envio

costo = float(input("Ingresa el coste de compra: "))
if costo >= 800:
 print("el envio es gratis")

if costo <= 799 + 150:
 print("se le agregara 150 pesos")
```
