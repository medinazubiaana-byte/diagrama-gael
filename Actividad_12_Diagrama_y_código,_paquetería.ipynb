#### Genera un codigo para una empresa de empaqueria. el programa debe pedir el peso del paque en kilos, la distancia del envio en kilometros y preguntar si el paquete es fragil. Si el peso es menor o igual a 5kg, el costo base es $100$. Si pesa mas de 5 kg, el costo base es de $200$.  Si la distancia es mayor a 100 km, se pagara $80$. Si el paquete es fragil, se agregan $50 por manejo especial. Mostrar el costo base, cargos extra y total final.

```python
peso_paquete = float(input("Ingresa el peso del paquete: "))
envio_kilometros = float(input("Ingresa cuantos kilometros de envio seran: "))
paquete_frajil = input("¿Su paquete es frajil?(si/no):")

if peso_paquete <= 5:
  costo_base = 100
elif peso_paquete > 5:
  costo_base = 200

cargos_extra = 0

# Costo por distancia
if envio_kilometros > 100:
  cargos_extra += 80

# Costo por fragilidad
if paquete_frajil.lower() == "si":
  cargos_extra += 50

total_final = costo_base + cargos_extra

print(f"Costo base: ${costo_base}")
print(f"Cargos extra: ${cargos_extra}")
print(f"Total final: ${total_final}")
```
