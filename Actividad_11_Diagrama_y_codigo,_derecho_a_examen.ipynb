#### Genera un diagrama de flujo que ayude al profesor a saber que alumno tendra el derecho de tener examen o no con las asistencias de las clases del todo el semestre, si no llega al 80% de las asistencias no tendra derecho a examen.

```python
total_clases = float(input("Ingresa el numero de clases totales: "))
asistio_alumno = float(input("Ingresa las asistencias del alumno: "))

if total_clases == 0:
  print("No se pueden calcular las asistencias si no hay clases totales.")
else:
  porcentaje_asistencia = (asistio_alumno / total_clases) * 100

  if porcentaje_asistencia >= 80:
    print("Tiene derecho a examen")
  else:
    print("No tiene derecho a examen")
```
